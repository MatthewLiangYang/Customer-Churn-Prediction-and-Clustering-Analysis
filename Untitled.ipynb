# Customer Churn Prediction and Clustering Analysis

**Background**:  

There is intense competition among online video service providers. To increase revenue, businesses need more subscribers, but retaining existing customers is more important than acquiring new ones. Therefore, businesses want to identify which customers are likely to cancel their subscriptions, i.e., customer churn.
If they can predict who is likely to churn, they may be able to retain them through promotional offers.

**Overview**:  

This dataset collects customer-related data regarding video service usage and labels whether a customer has churned or not.
Based on this dataset, this project will build a KNN classification model to predict customer churn. Additionally, all customers labeled as "churned" will undergo K-Means clustering analysis. Using pyecharts, we will visualize the results and finally propose targeted strategies for different types of churned customers.

### KMeans Clustering Analysis

```python
# Extract the dataset of churned customer groups
churn = data[data["Churn Status"] == 1]
churn = churn.reset_index(drop=True)
churn.head()
```

```python
churn.describe()
```

Calculation Formula:

C (Contract Status) = Remaining Contract Period

M (Spending Amount) = Average Download Volume + Average Upload Volume + Average Bill Amount

A (Activity Level) = If_TV + If_Movie_Package

S (Satisfaction Level) = Number of Complaints + Number of Restrictions

L (Service Duration) = Customer Tenure

```python
cal = {
    "C": churn["Remaining Contract Period"]
    ,"M": churn["Average Download Volume"] + churn["Average Upload Volume"] + churn["Average Bill Amount"]
    ,"A": churn["If_TV"] + churn["If_Movie_Package"]
    ,"S": churn["Number of Complaints"] + churn["Number of Restrictions"]
    ,"L": churn["Customer Tenure"]
}

data_class = pd.DataFrame(data=cal,columns=["C","M","A","S","L"])
data_class
```

```python
from sklearn.preprocessing import scale

# Standardization processing
data_class = pd.DataFrame(scale(data_class),
                          columns=["C", "M", "A", "S", "L"])
data_class_ID = pd.concat([churn[["Customer ID"]], data_class], axis=1)
data_class
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

np.random.seed(888)

# Initialize lists
score, inertia = [], []
nums = range(2, 10)

# Iterate through multiple possible candidate cluster numbers
for n_clusters in nums:
    kmeans = KMeans(n_clusters=n_clusters, random_state=888)
    kmeans.fit(data_class)
    inertia.append(kmeans.inertia_)
    score.append(silhouette_score(data_class, kmeans.labels_, sample_size=128, metric='euclidean'))

# Set plot style
plt.figure(figsize=(12, 4))
colors = ['#1f77b4', '#ff7f0e']

# Plot SSE
plt.subplot(1, 2, 1)
plt.plot(nums, inertia, marker='o', color=colors[0], linestyle='--', linewidth=2, markersize=8)
plt.xlabel('K', fontsize=12)
plt.ylabel('SSE', fontsize=12)
plt.title('SSE for Different Number of Clusters', fontsize=14, fontweight='bold')
plt.grid(linestyle=':', color='gray', alpha=0.6)

# Plot Silhouette Score
plt.subplot(1, 2, 2)
plt.plot(nums, score, marker='s', color=colors[1], linestyle='-.', linewidth=2, markersize=8)
plt.xlabel('K', fontsize=12)
plt.ylabel('Silhouette Score', fontsize=12)
plt.title('Silhouette Score for Different Number of Clusters', fontsize=14, fontweight='bold')
plt.grid(linestyle=':', color='gray', alpha=0.6)

# Set overall layout
plt.tight_layout()
plt.show()
```

```python
# K-Means Clustering
k = 5
model_KM = KMeans(n_clusters=k)
model_KM.fit(data_class)

print(model_KM.labels_)
```

```python
# View cluster centers
centers = pd.DataFrame(model_KM.cluster_centers_,
                       columns=data_class.columns)

# View clustering labels
labels = pd.DataFrame(model_KM.labels_, columns=["Cluster Label"])
```

```python
# View the number of customers in each cluster
labels_count = labels["Cluster Label"].value_counts().reset_index()
labels_count.columns = ["Cluster Label", "Number of Customers"]

labels_count = labels_count.sort_values("Cluster Label").reset_index(drop=True)

# View the number of customers in 5 clusters + five key dimensions
final = pd.concat([labels_count, centers], axis=1)
final["Cluster Label"] = ["class-1", "class-2", "class-3", "class-4", "class-5"]
final
```

```python
# Extract data and sort by the number of customers in descending order
pie_data = final.iloc[:, :2].sort_values("Number of Customers", ascending=False)
x_pie, y_pie = pie_data["Cluster Label"].tolist(), pie_data["Number of Customers"].tolist()

# Plot
pie = (
    Pie(init_opts=opts.InitOpts(width="500px", height="600px", theme="roma", bg_color="rgba(32,32,32,1)"))
    .add(
        "Churned Customer Clusters",
        [list(z) for z in zip(x_pie, y_pie)],
        radius=["20%", "80%"],
        center=["50%", "55%"]
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title="Number of Customers in Each Churned Cluster", pos_left="center", pos_top="5%",
            title_textstyle_opts=opts.TextStyleOpts(color="white", font_family="Microsoft YaHei", font_size=20)
        ),
        legend_opts=opts.LegendOpts(
            orient="vertical", type_="scroll",
            pos_right="5%", pos_top="5%",
            legend_icon="triangle", item_width=12,
            item_height=12,
            textstyle_opts=opts.TextStyleOpts(color="white")
        )
    )
    .set_series_opts(
        label_opts=opts.LabelOpts(position="inside", formatter="{b}:{c}")
    )
)

pie.render("pie_chart.html")
```

```python
from pyecharts import options as opts
from pyecharts.charts import Radar

# Get maximum and minimum values
value_range = {'min_': centers.values.min() - 0.1, 'max_': centers.values.max() + 0.1}

# Background grid line colors
split_line_style = {
    'color': ['rgba(255, 255, 255, 0.1)', 'rgba(255, 255, 255, 0.15)', 'rgba(255, 255, 255, 0.2)', 
              'rgba(255, 255, 255, 0.3)', 'rgba(255, 255, 255, 0.4)', 'rgba(255, 255, 255, 0.5)'][::-1]
}

# Indicator names and value ranges
indicators = ["C", "M", "A", "S", "L"]
schema = [opts.RadarIndicatorItem(name=ind, **value_range) for ind in indicators]

# Initialize radar chart
radar = (
    Radar(init_opts=opts.InitOpts(width="800px", height="800px", theme="macarons", bg_color="rgba(30,30,40,0.9)"))
    .add_schema(
        shape="circle",
        splitline_opt=opts.SplitLineOpts(is_show=True, linestyle_opts=split_line_style),
        schema=schema,
        textstyle_opts=opts.TextStyleOpts(color="#FFFFFF", font_size=14)  # Indicator text style
    )
)

# Classification data and color configuration
class_data = [
    ("Class 1", class_1, "#4682B4"),  # Steel Blue
    ("Class 2", class_2, "#8A2BE2"),  # Blue Violet
    ("Class 3", class_3, "#FF7F50"),  # Coral
    ("Class 4", class_4, "#32CD32"),  # Lime Green
    ("Class 5", class_5, "#FFD700"),  # Gold
]

# Add each class data and apply gradient colors
for label, data, color in class_data:
    radar.add(
        label, data,
        linestyle_opts=opts.LineStyleOpts(color=color, width=1.5, opacity=0.85),
        areastyle_opts=opts.AreaStyleOpts(color=color, opacity=0.2),
        label_opts=opts.LabelOpts(is_show=False)
    )

# Global configuration optimization
radar.set_global_opts(
    title_opts=opts.TitleOpts(
        title="Radar Chart of Churned Customer Clusters",
        subtitle="Based on K-Means Analysis",
        pos_left="center",
        pos_top="2%",
        title_textstyle_opts=opts.TextStyleOpts(color="white", font_family="Microsoft YaHei", font_size=20, font_weight="bold"),
        subtitle_textstyle_opts=opts.TextStyleOpts(color="#CCCCCC", font_family="Microsoft YaHei", font_size=14)
    ),
    legend_opts=opts.LegendOpts(
        is_show=True,
        pos_right="10%",
        pos_top="10%",
        orient="vertical",
        legend_icon="circle",
        item_width=14,
        item_height=14,
        type_="scroll",
        textstyle_opts=opts.TextStyleOpts(color="white", font_family="Microsoft YaHei", font_size=12)
    )
)

# Render and display
radar.render("rader_chart.html")
```
